```python
from PIL import Image
import numpy as np
import os, os.path, time
import sys
import csv
from numpy import zeros
import random
# Much of this file comes from Keras demo code
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import regularizers

# Just in case the Keras defaults aren't as we expect
from keras import backend as K_backend
K_backend.set_image_data_format('channels_last')
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot


```

```python
# 256*256


img_rows,img_cols=32,32

#read files
fileList=[]
rootdir='Pan_img'
for root, dirs, files in os.walk(rootdir, topdown=False):
    for name in files:
        #print(name)
        if name.endswith('.jpg'):
            fullName = os.path.join(root, name)
            fileList.append(fullName)
            
dataset = []

for file in fileList:
    print(file)
    img_file = Image.open(file)
    form = img_file.format
    print(form)
    img_file = img_file.resize((img_rows,img_cols), Image.ANTIALIAS)
    #form = img_file.format
    #print (form)
    arr=zeros(2)
    arr[1]=1
    # Make image Greyscale
    img_grey = img_file.convert('L')
        
    # Save Greyscale values
    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    dataset.append([value,arr])
    
fileList=[]
rootdir='aadhaar_img'
for root, dirs, files in os.walk(rootdir, topdown=False):
    for name in files:
        #print(name)
        if name.endswith('.jpg'):
            fullName = os.path.join(root, name)
            fileList.append(fullName)
            
for file in fileList:
    print(file)
    img_file = Image.open(file)
    form = img_file.format
    print(form)
    img_file = img_file.resize((img_rows,img_cols), Image.ANTIALIAS)
    #form = img_file.format
    #print (form)
    arr=zeros(2)
    arr[0]=1
    # Make image Greyscale
    img_grey = img_file.convert('L')
        
    # Save Greyscale values
    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    dataset.append([value,arr])
    
```

```python
random.shuffle(dataset)    
print(len(dataset))

# split into train and test sets
train_size = int(len(dataset) * 0.7) 
test_size = len(dataset) - train_size
train, test = dataset[0:train_size][:], dataset[train_size:len(dataset)][:]

from numpy import array
train=array(train)
test=array(test)
#print(type(train))

x_train=[]
y_train=[]
x_test=[]
y_test=[]

for b in train:
    x_train.append(b[0])
    y_train.append(b[1])
    
for c in test:
    x_test.append(c[0])
    y_test.append(c[1])
    
from numpy import array
x_train=array(x_train)
y_train=array(y_train)
x_test=array(x_test)
y_test=array(y_test)

#check shape of all training datsets is same
#for m in x_train:
    #print(m.shape) 
    
#print(type(x_train))

x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
#print('x_train shape:', x_train.shape)
#print(x_train.shape[0], 'train samples')
#print(x_test.shape[0], 'test samples')

#print(y_train.shape, 'train samples') 
#print(type(y_train))

```

```python
datagen = ImageDataGenerator(horizontal_flip=True, vertical_flip=False,validation_split=0.25)

# fit parameters from data
datagen.fit(x_train)

# configure batch size and retrieve one batch of images
for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=1):
# create a grid of 3x3 images
    for i in range(0, 1):
        pyplot.imshow(X_batch[i].reshape(32,32), cmap=pyplot.get_cmap('gray'))
# show the plot
    pyplot.show()
    break
    
batch_size = 4
steps_per_epoch=(0.25*len(x_train))//batch_size
validation_steps=(0.75*len(x_train))//batch_size 
print(steps_per_epoch)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
batch_size = 4
num_classes = 2
epochs = 75
input_shape=(img_rows,img_cols,1)

pat = 30 #this is the number of epochs with no improvment after which the training will stop
#mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
#es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=pat)

#model 
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.4))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

#train model
history = model.fit_generator(datagen.flow(x_train, y_train, batch_size=batch_size, subset='training'),
                    steps_per_epoch=steps_per_epoch,
                    epochs=epochs,
                    validation_data=datagen.flow(x_train, y_train, batch_size=batch_size, subset='validation'),
                    validation_steps=validation_steps)


#evaluate model 
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# Plot training
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('prediction')
plt.xlabel('epoch')
plt.legend(['loss', 'validation loss'], loc='upper right')
font = {'weight' : 'bold',
        'size'   : 18}
plt.rc('font', **font)
plt.grid()
plt.show()
```
